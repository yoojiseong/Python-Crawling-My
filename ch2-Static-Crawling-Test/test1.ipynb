```python
#bs4 (BeautifulSoup)는
# 웹 페이지의 HTML이나 XML 문서로부터
# 데이터를 쉽게 추출하기 위한 파이썬 라이브러리입니다.
# 복잡한 웹 페이지 소스 코드를 마치
# 잘 정리된 책의 목차처럼 다룰 수 있게 해주는 강력한 도구이죠.

## 기본 개념
# 웹 페이지는 수많은 태그(tag)들로 구성된 복잡한 구조를 가집니다.
# BeautifulSoup은 이 지저분하고 복잡한 소스 코드를 파이썬이
# 이해하기 쉬운 객체 구조(Parse Tree)로 변환해 줍니다.
# 이 구조를 통해 우리는 특정 태그나 원하는 정보에 쉽게 접근하여 데이터를 추출할 수 있습니다.
#
# 예를 들어,
# 웹 페이지를 '집'이라고 비유한다면,
# BeautifulSoup은 각 방(태그)의 주소를 알려주고,
# 방 안에 있는 물건(내용, 속성)을 쉽게 가져올 수 있도록
# 도와주는 '설계도'와 같습니다.

#기본 문법
# BeautifulSoup의 핵심 문법
# 몇 가지만 알면 기본적인 데이터 추출이 가능합니다.
#
# 1. BeautifulSoup 객체 생성
# 먼저 HTML 데이터를 파싱하여 soup 객체를 생성해야 합니다.
from bs4 import BeautifulSoup

# 분석하고 싶은 HTML 데이터
html_doc = "<html><head><title>The Dormouse's story</title></head><body>...</body></html>"

# BeautifulSoup 객체 생성 (HTML 데이터, '사용할 파서')
soup = BeautifulSoup(html_doc, 'html.parser')


# 2. 태그(Tag) 찾기: find()와 find_all()
# soup.find('태그명'): 조건에 맞는 가장 첫 번째 태그 하나를 찾습니다.
# soup.find_all('태그명'): 조건에 맞는 모든 태그를 찾아 리스트(List) 형태로 반환합니다.
# 가장 먼저 나오는 <p> 태그 하나를 찾기
p_tag = soup.find('p')
# 모든 <li> 태그를 찾기
li_tags = soup.find_all('li')

# 3 태그의 내용(Text) 가져오기: .text
# 태그 객체에서 .text 속성을 사용하면
# 태그 안의 모든 텍스트 내용을 가져올 수 있습니다.
# <title> 태그의 텍스트 가져오기
title_text = soup.find('title').text
print(title_text)  # 출력: The Dormouse's story

# 4. 태그의 속성(Attribute) 값 가져오기
# 태그의 속성(예: 링크 주소 href, 이미지 주소 src)은
# 딕셔너리처럼 대괄호 []를 이용해 접근합니다.
# <a> 태그를 찾고, 그 안의 href 속성 값을 가져오기
a_tag = soup.find('a')
link_url = a_tag['href']
print(link_url)

#  5 CSS 선택자로 찾기: select()와 select_one()
# CSS 선택자 문법에 익숙하다면 select()가 훨씬 직관적이고 강력합니다.
#
# soup.select_one('선택자'): CSS 선택자에 맞는 첫 번째 요소를 찾습니다.
#
# soup.select('선택자'): CSS 선택자에 맞는 모든 요소를 리스트로 찾습니다.
```

```python
import bs4
webPage = open(r'./Sample03.html','rt',encoding='utf-8').read()
bsObject2 = bs4.BeautifulSoup(webPage, "html.parser")
print(bsObject2)
```

```python
tag_id_myId1 = bsObject2.find('div', {'id': 'myId1'})
print(tag_id_myId1)
```

```python
tag_class_myClass1 = bsObject2.find('div', {'class': 'myClass1'})
print(tag_class_myClass1)
```

```python
ul_value = bsObject2.find('ul', {'class': 'myClass2'})
print(ul_value)
```

```python
tag_div_myClass1 = bsObject2.findAll('div', {'class': 'myClass1'})
print(tag_div_myClass1)
```

```python
li_list_myClass3 = bsObject2.findAll('li', {'class': 'myClass3'})
print(li_list_myClass3)
```

```python
a_list = bsObject2.findAll('a')
print(a_list)
```

```python
for aTag in a_list:
    print(aTag['href'])
```

```python
tag_li_all = bsObject2.findAll('li')
for tag_li in tag_li_all:
    print(tag_li.text)
```
