```python
# 알라딘 중고서점, 신간 베스트, 자기 계발,
# 현재 랭킹 1~50, 나와있고,
# 가져올 데이터
#1) 책 제목, 2) 정가 3) 지은이

# https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=TodayHot&BranchType=1&CID=336
```

```python
import bs4
import urllib.request
import csv
import json
import pymysql

aladinUrl = "https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=TodayHot&BranchType=1&CID=336"
htmlObject = urllib.request.urlopen(aladinUrl)
webPage = htmlObject.read()
bsObject = bs4.BeautifulSoup(webPage, 'html.parser')
print(bsObject)
```

```python
#책 정보 가져오기 확인 작업.
# 책 하나당. ss_book_list 2개이고,
# 우리가 원하는 정보는 첫번째 ss_book_list
# 책제목. a -> class = bo3, 50개,
# ss_book_list -> ul -> li(2번째 ) -> a -> class = bo3
# 정가
# ss_book_list -> ul -> li(4번째 ) -> span : 정가
# 정가,
# ul -> 2가지, 빨간 광고 문구 o li 5개 중 -> 4번째 li -> span
# ul -> 2가지, 빨간 광고 문구 x li 4개 중 -> 3번째 li -> span
# 지은이
# ss_book_list -> ul -> li(3번째 ) -> a 태그 2개중 첫번째 : 지은이
# ul -> 2가지, 빨간 광고 문구 o li 5개 중 -> 3번째 li -> a
# ul -> 2가지, 빨간 광고 문구 x li 4개 중 -> 2번째 li -> a
book_all_100 = bsObject.findAll('div',{'class':'ss_book_list'})
# print(book_all_100)
# print(len(book_all_100))
# enumerate => 리스트를 꺼낼때 인덱스도 같이 꺼낸다.
for index, book in enumerate(book_all_100):
    if index % 2 == 0:
        print(f"index : {index}")
        # print(f"index : {index}\n book{[index]}{book}")
        book_ul = book.find('ul')
        # book_ul_li_list = book_ul.findAll('li')
        book_title = book_ul.find('a', {'class':'bo3'})
        # print(book_ul_li_list)
        print(f"책 제목 : {book_title.text}")

        book_price = ""
        book_author = ""
        book_ul_li_list = book_ul.findAll('li')
        if len(book_ul_li_list) == 5:
            book_price = book_ul_li_list[3].find('span').text
            book_author = book_ul_li_list[2].find('a').text

        elif len(book_ul_li_list) == 4:
            book_price = book_ul_li_list[2].find('span').text
            book_author = book_ul_li_list[1].find('a').text
        print(f"책 가격 : {book_price}")
        print(f"책 저자 : {book_author}")
        # print(book_price)
        # print(book_author)
```

```python
#csv, json 파일 쓰기 작업.
# book_all_100
book_data = []
book_all_100 = bsObject.findAll('div',{'class':'ss_book_list'})
for index, book in enumerate(book_all_100):
    if index % 2 == 0:
        book_ul = book.find('ul')
        book_title = book_ul.find('a', {'class':'bo3'}).text
        book_price = ""
        book_author = ""
        book_ul_li_list = book_ul.findAll('li')
        if len(book_ul_li_list) == 5:
            book_price = book_ul_li_list[3].find('span').text
            book_author = book_ul_li_list[2].find('a').text

        elif len(book_ul_li_list) == 4:
            book_price = book_ul_li_list[2].find('span').text
            book_author = book_ul_li_list[1].find('a').text
        # book_data.append({"title" : book_title, "price" : book_price, "author" : book_author})
        book_data.append((book_title, book_price, book_author))

#csv 저장,
with open(r"./book_all_50.csv", "w",newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["title", "price", "author"])
    writer.writerows(book_data)

print(f"\n✅ CSV 파일 저장 완료: book_all_50.csv")
```

```python
#json
with open(r"./book_all_50.json", "w", encoding="utf-8") as f:
    json.dump(book_data, f, ensure_ascii=False, indent=4)

print(f"✅ JSON 파일 저장 완료: book_all_50.json")
```

```python
#마리아디비에 쓰기 작업
# id : webuser, pw : webuser, db:webdb
# 샘플 테이블
# CREATE TABLE booksAladin (
#     id INT AUTO_INCREMENT PRIMARY KEY,
#     title VARCHAR(255) NOT NULL,
#     price VARCHAR(50),
#     author VARCHAR(100)
# );

# db 연결 작업,
db = pymysql.connect(host="localhost", user="webuser", passwd="webuser", db="webdb")

cursor = db.cursor()

query = "INSERT INTO booksAladin (title, price, author) VALUES (%s, %s, %s)"

try:
    cursor.executemany(query, book_data)
    db.commit()
    print(f"DB 저장 완료")
except Exception as e:
    db.rollback()
    print(f"DB 저장 실패 : {e}")
cursor.close()
db.close()
```
